````python
Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.

 Here's a Python code that creates a `Vehicle` class with an `__init__` method having instance variables `name_of_vehicle`, `max_speed`, and `average_of_vehicle`:

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

# Example usage:
car = Vehicle("Toyota Camry", 180, 25)
print("Name of Vehicle:", car.name_of_vehicle)
print("Max Speed:", car.max_speed)
print("Average of Vehicle:", car.average_of_vehicle)
```

In this code:
- We define a class named `Vehicle`.
- The `__init__` method is a constructor that initializes the instance variables `name_of_vehicle`, `max_speed`, and `average_of_vehicle` with the values passed as arguments.
- Example usage demonstrates creating an instance of the `Vehicle` class and accessing its attributes.
````

````python
Q2.  Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.

 Here's the Python code implementing the `Car` class as a child class of the `Vehicle` class with a method named `seating_capacity`:

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

class Car(Vehicle):
    def seating_capacity(self, capacity):
        return f"The {self.name_of_vehicle} has a seating capacity of {capacity}."

# Example usage:
car = Car("Toyota Camry", 180, 25)
print(car.seating_capacity(5))
```

In this code:
- We define a child class named `Car` that inherits from the `Vehicle` class.
- The `seating_capacity` method is defined within the `Car` class. It takes a parameter `capacity` representing the seating capacity and returns a formatted string containing the name of the vehicle and its seating capacity.
- Example usage demonstrates creating an instance of the `Car` class and calling the `seating_capacity` method.
````

````python
Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

Multiple inheritance is a feature in object-oriented programming languages where a class can inherit attributes and methods from more than one parent class. In Python, a class can inherit from multiple parent classes by listing them in parentheses after the class name.

Here's an example Python code to demonstrate multiple inheritance:

```python
class A:
    def method_A(self):
        print("Method A from class A")

class B:
    def method_B(self):
        print("Method B from class B")

class C(A, B):
    def method_C(self):
        print("Method C from class C")

# Create an instance of class C
obj = C()

# Call methods from class A
obj.method_A()

# Call methods from class B
obj.method_B()

# Call method from class C
obj.method_C()
```

In this code:
- Class `A` defines a method `method_A`.
- Class `B` defines a method `method_B`.
- Class `C` inherits from both classes `A` and `B`, thereby exhibiting multiple inheritance.
- Instances of class `C` have access to methods from both classes `A` and `B`, as well as their own methods defined in class `C`.

Output:
```
Method A from class A
Method B from class B
Method C from class C
```

This demonstrates how a class can inherit methods and attributes from multiple parent classes in Python. Multiple inheritance can be useful but should be used carefully to avoid ambiguity and complexity in the code.
````

````python
Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.

In Python, getters and setters are methods used to access and modify the attributes (variables) of a class. They provide a way to encapsulate the access to class attributes, allowing for better control over how attributes are accessed and modified.

- **Getter Method**: A getter method is used to retrieve the value of an attribute.
- **Setter Method**: A setter method is used to modify the value of an attribute.

Here's an example of a class with getter and setter methods in Python:

```python
class Student:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    # Getter method for name attribute
    def get_name(self):
        return self._name

    # Setter method for name attribute
    def set_name(self, name):
        self._name = name

    # Getter method for age attribute
    def get_age(self):
        return self._age

    # Setter method for age attribute
    def set_age(self, age):
        if age >= 0:  # Adding validation for age
            self._age = age
        else:
            print("Age cannot be negative.")

# Create an instance of the Student class
student1 = Student("Alice", 20)

# Get the current name and age
print("Name:", student1.get_name())
print("Age:", student1.get_age())

# Set a new name and age using setter methods
student1.set_name("Bob")
student1.set_age(25)

# Get the updated name and age
print("Updated Name:", student1.get_name())
print("Updated Age:", student1.get_age())
```

In this code:
- We define a `Student` class with `name` and `age` attributes.
- Getter methods (`get_name` and `get_age`) are defined to retrieve the values of the attributes.
- Setter methods (`set_name` and `set_age`) are defined to modify the values of the attributes.
- We create an instance of the `Student` class and use the getter and setter methods to access and modify the attributes.

Using getter and setter methods provides control over attribute access and modification, allowing for validation, logging, or other operations to be performed. Additionally, using getter and setter methods helps achieve encapsulation, as the internal representation of the class attributes can be hidden from the outside world.
````

````python
Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

Method overriding in Python is a feature that allows a subclass to provide a specific implementation of a method that is already defined in its superclass. This means that a subclass can redefine a method of its superclass with the same name and signature.

Here's a Python code example demonstrating method overriding:

```python
class Animal:
    def make_sound(self):
        print("Generic animal sound")

class Dog(Animal):
    def make_sound(self):
        print("Bark")

class Cat(Animal):
    def make_sound(self):
        print("Meow")

# Create instances of Dog and Cat
dog = Dog()
cat = Cat()

# Call the make_sound method of each instance
dog.make_sound()  # Output: Bark
cat.make_sound()  # Output: Meow
```

In this code:
- We define a superclass `Animal` with a method `make_sound` that prints a generic animal sound.
- We define two subclasses `Dog` and `Cat` that inherit from `Animal`.
- Each subclass overrides the `make_sound` method with its own implementation (`Bark` for `Dog` and `Meow` for `Cat`).
- When we create instances of `Dog` and `Cat` and call the `make_sound` method on each instance, Python resolves the method call to the overridden method in the respective subclass.

Method overriding allows subclasses to provide specialized behavior while still maintaining a common interface defined by the superclass. It is a fundamental concept in object-oriented programming that promotes code reuse and polymorphism.
````
